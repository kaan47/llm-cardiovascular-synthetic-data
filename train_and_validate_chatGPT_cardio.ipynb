```python
# =================================================================
# Optimized Stacking Classifier - Final Performance Evaluation
# =================================================================

import pandas as pd
import numpy as np
from sklearn.ensemble import (AdaBoostClassifier, GradientBoostingClassifier, 
                              RandomForestClassifier, HistGradientBoostingClassifier)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from mlxtend.classifier import StackingCVClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# ----------------------------
# 1. Data Loading & Preprocessing
# ----------------------------
train_df = pd.read_csv("chatGPT_cardio.csv")
test_df = pd.read_csv("real_data.csv")

train_df.columns = [col.replace(" ", "_") for col in train_df.columns]
test_df.columns = [col.replace(" ", "_") for col in test_df.columns]

numerical_cols = ['age', 'resting_bp_s', 'cholesterol', 'max_heart_rate', 'oldpeak']
scaler = StandardScaler()

train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])
test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])

X_train = train_df.drop(columns=['target']).values # Converting to numpy for mlxtend
y_train = train_df['target'].values
X_test = test_df.drop(columns=['target']).values
y_test = test_df['target'].values

# ----------------------------
# 2. Define Optimized Base Learners
# ----------------------------
# These parameters are taken from your provided best search results
ada = AdaBoostClassifier(learning_rate=1, n_estimators=150, random_state=42)

gbm = GradientBoostingClassifier(learning_rate=0.1, max_depth=3, n_estimators=200, 
                                 subsample=1.0, random_state=42)

xgb = XGBClassifier(learning_rate=0.01, max_depth=3, n_estimators=100, 
                    subsample=0.8, eval_metric="logloss", random_state=42)

lgb = LGBMClassifier(learning_rate=0.1, max_depth=3, n_estimators=100, 
                     subsample=0.8, random_state=42, verbose=-1)

rf = RandomForestClassifier(max_depth=None, min_samples_leaf=1, min_samples_split=2, 
                            n_estimators=100, random_state=42)

s_gbm = GradientBoostingClassifier(learning_rate=0.05, max_depth=5, n_estimators=50, 
                                   subsample=0.6, random_state=42)

ada_rf = AdaBoostClassifier(estimator=RandomForestClassifier(max_depth=10, n_estimators=100, random_state=42), 
                            learning_rate=0.01, n_estimators=50, random_state=42)

# ----------------------------
# 3. Define Meta Learner & Build Stack
# ----------------------------
meta_learner = HistGradientBoostingClassifier(random_state=42)

stacking_clf = StackingCVClassifier(
    classifiers=[ada, gbm, xgb, lgb, rf, s_gbm, ada_rf],
    meta_classifier=meta_learner,
    use_probas=True,
    cv=5,
    random_state=42
)

# Train the stack on synthetic data
stacking_clf.fit(X_train, y_train)

# ----------------------------
# 4. Final Evaluation on Real Data
# ----------------------------
y_pred_stack = stacking_clf.predict(X_test)
y_probs_stack = stacking_clf.predict_proba(X_test)[:, 1]

final_acc = accuracy_score(y_test, y_pred_stack)
final_auc = roc_auc_score(y_test, y_probs_stack)

print(f"Final Stacking Accuracy (Real Data): {final_acc:.4f}")
print(f"Final Stacking AUC Score:           {final_auc:.4f}")
print("\n--- Detailed Classification Report (Stacking) ---")
print(classification_report(y_test, y_pred_stack))
```

```python
# =================================================================
# Manual Blending (Out-of-Fold) - Final Performance Evaluation
# =================================================================

import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import (AdaBoostClassifier, GradientBoostingClassifier, 
                              RandomForestClassifier)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# ----------------------------
# 1. Data Loading & Preprocessing
# ----------------------------
train_df = pd.read_csv("chatGPT_cardio.csv")
test_df = pd.read_csv("real_data.csv")

train_df.columns = [col.replace(" ", "_") for col in train_df.columns]
test_df.columns = [col.replace(" ", "_") for col in test_df.columns]

numerical_cols = ['age', 'resting_bp_s', 'cholesterol', 'max_heart_rate', 'oldpeak']
scaler = StandardScaler()

train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])
test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])

X_train = train_df.drop(columns=['target']).values
y_train = train_df['target'].values
X_test = test_df.drop(columns=['target']).values
y_test = test_df['target'].values

# ----------------------------
# 2. Define Optimized Base Models
# ----------------------------
base_models = [
    ('ada', AdaBoostClassifier(learning_rate=1, n_estimators=150, random_state=42)),
    ('gbm', GradientBoostingClassifier(learning_rate=0.1, max_depth=3, n_estimators=200, subsample=1.0, random_state=42)),
    ('xgb', XGBClassifier(learning_rate=0.01, max_depth=3, n_estimators=100, subsample=0.8, eval_metric="logloss", random_state=42)),
    ('lgb', LGBMClassifier(learning_rate=0.1, max_depth=3, n_estimators=100, subsample=0.8, random_state=42, verbose=-1)),
    ('rf', RandomForestClassifier(max_depth=None, min_samples_leaf=1, min_samples_split=2, n_estimators=100, random_state=42)),
    ('s_gbm', GradientBoostingClassifier(learning_rate=0.05, max_depth=5, n_estimators=50, subsample=0.6, random_state=42)),
    ('ada_rf', AdaBoostClassifier(estimator=RandomForestClassifier(max_depth=10, n_estimators=100, random_state=42), 
                                  learning_rate=0.01, n_estimators=50, random_state=42))
]

# ----------------------------
# 3. Generate Base Model Predictions (5-Fold Blending)
# ----------------------------
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
# Stores OOF predictions to train the Meta-Learner
base_predictions_proba = np.zeros((X_train.shape[0], len(base_models)))
# Stores test set predictions for each fold
test_fold_predictions = np.zeros((5, X_test.shape[0], len(base_models)))

for i, (train_index, val_index) in enumerate(kf.split(X_train, y_train)):
    X_tr, X_val = X_train[train_index], X_train[val_index]
    y_tr, y_val = y_train[train_index], y_train[val_index]
    
    for j, (name, model) in enumerate(base_models):
        # Train base model on the current fold
        model.fit(X_tr, y_tr)
        
        # Predict probabilities on validation fold (Out-of-Fold)
        base_predictions_proba[val_index, j] = model.predict_proba(X_val)[:, 1]
        
        # Predict probabilities on the actual test set (real_data.csv)
        test_fold_predictions[i, :, j] = model.predict_proba(X_test)[:, 1]

# Average the test set predictions across all 5 folds
test_predictions_proba_avg = test_fold_predictions.mean(axis=0)

# ----------------------------
# 4. Train Meta-Model & Final Prediction
# ----------------------------
# Meta-learner is trained on the OOF predictions from the synthetic data
meta_model = LogisticRegression()
meta_model.fit(base_predictions_proba, y_train)

# Final prediction probabilities on real_data.csv
final_probs = meta_model.predict_proba(test_predictions_proba_avg)[:, 1]
# Final binary classes based on 0.5 threshold
final_preds = (final_probs >= 0.5).astype(int)

# ----------------------------
# 5. Final Output
# ----------------------------
auc_score = roc_auc_score(y_test, final_probs)
acc_score = accuracy_score(y_test, final_preds)

print(f"\nFinal Blending Accuracy (Real Data): {acc_score:.4f}")
print(f"Final Blending AUC Score:            {auc_score:.4f}")
print("\n--- Detailed Classification Report ---")
print(classification_report(y_test, final_preds))
```
